```python
for col in df.columns:
    if df[col].dtypes == int or df[col].dtypes == float:
        df2 = df[col]
        df2 = df2.dropna()
        
        if abs(np.mean(df2) - np.median(df2)) <= np.std(df2):
            df[col].fillna(np.mean(df2), inplace = True)
        else:
            df[col].fillna(np.median(df2), inplace = True)
```

```python
for colum in df.columns:
    print(df[colum].dtypes)
    if ((df[colum].dtypes == object) or (df[colum].dtypes == str)):
        df3 = df[colum]
        df3 = df3.dropna()
        print("A")
        unq_val = list(df3.unique())
        print("unq_val: ", unq_val)
        if len(unq_val) <= 10:
            total_count = 0
            val = str
            for v in unq_val:
                if v != 'nan':
                    count = df3.loc[df3 == v]
                    print(count.shape[0])
                    count = int(count.shape[0])
                    
                    if count >= total_count:
                        total_count = count
                        val = v
                        
            df[colum].fillna(val, inplace = True)
            
        else:
            df.drop(colum, axis = 1, inplace = True)
            print("Dropped: ", colum)
```
